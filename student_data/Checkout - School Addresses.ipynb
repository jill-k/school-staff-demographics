```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
import re
import glob
import os
```

```python
base = 'http://profiles.doe.mass.edu/search/search_export.aspx?orgCode=&orgType=6,13&runOrgSearch=Y&searchType=ORG&leftNavId=11238&showEmail=N'
```

```python
url = base
req = requests.get(url)
html = req.content
soup = bs(html, 'html.parser')
```

```python
schools = soup.find_all("tr")
schools = [x.text for x in schools]
sch_df = pd.DataFrame([sub.split("\n") for sub in schools])
cols = ["Drop", "Org Name", "Org Code", "Org Type", "Function", "Contact", "Address 1", "Address 2", "Town", "State", "Zip Code", "Phone", "Fax", "Grades", "Drop 2"]

sch_df.columns = cols
sch_df = sch_df[1:]
sch_df.set_index(('Org Code'), inplace=True)
sch_df.drop(["Drop", "Drop 2"], axis = 1, inplace=True)

sch_df
```
