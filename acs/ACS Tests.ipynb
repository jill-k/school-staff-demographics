```python
import numpy as np
import pandas as pd
import requests
```

```python
def get_ma_data(variable):
  api_base = "https://api.census.gov/data/2015/acs5/profile?for=zip%20code%20tabulation%20area:*&get=NAME,"
  return requests.get(api_base + variable).json()
```

```python
ma_zip_codes = pd.read_csv("school_addresses.csv", encoding = "ISO-8859-1", converters={'Zip Code': str})
zip_list = list(set(ma_zip_codes['Zip Code'].tolist()))
```

```python
median_income = get_ma_data("DP03_0062E")
```

```python
df = []

for i in median_income:
    if i[2] in zip_list:
        df.append(i)
```

```python
median_income_df = pd.DataFrame(df, columns = ["Name", "Median Income", "Zip Code"])
```

```python
median_income_df
```

```python
perc_poverty_kids = get_ma_data("DP03_0120PE")
```

```python
df = []

for i in perc_poverty_kids:
    if i[2] in zip_list:
        df.append(i)

percent_poverty_kids_df = pd.DataFrame(df, columns = ["Name", "Percent Poverty", "Zip Code"])
```

```python
hs_grad = get_ma_data("DP02_0066PE")
```

```python
df = []

for i in hs_grad:
    if i[2] in zip_list:
        df.append(i)
```

```python
hs_grad_df = pd.DataFrame(df, columns = ["Name", "% HS Graduates", "Zip Code"])
```

```python
college_grad = get_ma_data("DP02_0067PE")
```

```python
df = []

for i in college_grad:
    if i[2] in zip_list:
        df.append(i)
```

```python
college_grad_df = pd.DataFrame(df, columns = ["Name", "% College Graduates", "Zip Code"])
```

```python
college_grad_df
```

```python
acs_all = pd.merge(median_income_df, percent_poverty_kids_df)
acs_all = pd.merge(acs_all, hs_grad_df)
acs_all = pd.merge(acs_all, college_grad_df)
```

```python
acs_all.to_csv("acs_indicators_by_zip.csv")
```
